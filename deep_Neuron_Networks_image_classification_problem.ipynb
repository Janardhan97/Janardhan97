```python
(trainX,trainY), (testX, testY) = tf.keras.datasets.mnist.load_data()
```

```python
trainY = tf.keras.utils.to_categorical(trainY, num_classes = 10)
testY = tf.keras.utils.to_categorical(testY , num_classes=10)
```

```python
#building dnn model multilayer precptron
```

```python
model1 = tf.keras.models.Sequential()

#reshape 2D to 1D - 28 *28 = 784

model1.add(tf.keras.layers.Reshape((784,), input_shape=(28,28,)))
```

```python
##normalize the dATA 

model1.add(tf.keras.layers.BatchNormalization())
```

```python
# add a  hidden layer

model1.add(tf.keras.layers.Dense(20,activation='relu'))
```

```python
model1.add(tf.keras.layers.Dense(20,activation='relu'))
```

```python
#output  
```

```python
model1.add(tf.keras.layers.Dense(10, activation='relu'))
```

```python
# compile

model1.compile(optimizer = 'sgd' , loss = 'categorical_crossentropy', metrics= ['accuracy'])
```
